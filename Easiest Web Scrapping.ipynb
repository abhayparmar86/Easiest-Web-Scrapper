Examples that I have implemented to scrap tables from multiple websites. You can change above code as per your requirements.

```python
# Example 1

import pandas as pd

for i in range(1,3):
  url = "http://www.tenders.tn.gov.in/innerpage.asp?choice=ot4&page="+str(i)
  tables = pd.read_html(url)
  # print(tables)
  Table = tables[9].iloc[:, 1:5]
  # print(Table)

  output_csv = "output"+str(i)+".csv"
  print(output_csv)

  Table.to_csv(output_csv, index=False)
```

```python
# Example 2

# Removed indexing in 1st row, and page details from last row
import pandas as pd

for i in range(1, 3):
    url = "http://www.tenders.tn.gov.in/innerpage.asp?choice=ot4&page=" + str(i)
    tables = pd.read_html(url)
    Table = tables[9].iloc[:-1, 1:5]

    if i == 1:
        output_csv = "output.csv"
        Table.to_csv(output_csv, index=False, header = False)
    else:
      Table = Table.iloc[1:-1]
      with open(output_csv, 'a', newline='') as file:
            Table.to_csv(file, index=False, header=False)
```
